```python
import matplotlib.pyplot as plt

lines = []
with open('results.txt') as f:
    lines = f.readlines()

print (lines[2])
count = 0


x = []
y = []


for i in range (0, len(lines)-3, 3):
    #print( f"{i+1}: {lines[i]}" )
    #print( f"{i+2}: {lines[i+1]}" )

    #print(int(lines[i+1])/100)

    x.append(int(lines[i]))
    y.append(int(lines[i+1])/100)


print (x)
print(y)


for line in lines:
    count += 1
    #print(f'line {count}: {line}')
    #print(f'line {count+1}: {lines[count]}')
    count += 2




plt.plot(x, y, 'bo-', label='line 1', linewidth=1)

plt.xscale('symlog')

#plt.axis([0, 33554432, 0, 100])
plt.ylabel("accuracy (%)")  




#plt.style.use('ggplot')

fig = plt.figure(figsize=(15, 4))
ax = fig.add_subplot(111)
ax.set_xscale("log")
ax.plot(x, y, 'bo-', label='line 1', linewidth=1)
ax.set_xlabel("Number of bit flips")
ax.set_ylabel("accuracy (%)")
ax.set_title("BER sensitivity")




x_ber = []

for i in range (len(x)):
    x_ber.append((x[i]/100000000))

fig_a = plt.figure(figsize=(15, 4))
ax_a = fig_a.add_subplot(111)
ax_a.set_xscale("log")
ax_a.plot(x_ber, y, 'ro-', label='line 1', linewidth=1)
ax_a.set_xlabel("Bit Error Rate")
ax_a.set_ylabel("accuracy (%)")
ax_a.set_title("BER sensitivity")




x_a = []

for i in range (len(x)):
    x_a.append((x[i]/47185919)*100)

fig_a = plt.figure(figsize=(15, 4))
ax_a = fig_a.add_subplot(111)
ax_a.set_xscale("linear")
ax_a.plot(x_a, y, 'go-', label='line 1', linewidth=1)
ax_a.set_xlabel("Portion of bit flips (%)")
ax_a.set_ylabel("accuracy (%)")
ax_a.set_title("BER sensitivity")


plt.show()
```
