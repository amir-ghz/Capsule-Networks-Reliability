```python
import matplotlib.pyplot as plt

lines = []
with open('results.txt') as f:
    lines = f.readlines()

print (lines[2])
count = 0


x = []
y = []


for i in range (0, len(lines)-3, 3):
    #print( f"{i+1}: {lines[i]}" )
    #print( f"{i+2}: {lines[i+1]}" )

    #print(int(lines[i+1])/100)

    x.append(int(lines[i]))
    y.append(int(lines[i+1])/100)


print (x)
print(y)


for line in lines:
    count += 1
    #print(f'line {count}: {line}')
    #print(f'line {count+1}: {lines[count]}')
    count += 2




plt.plot(x, y, 'bo-', label='line 1', linewidth=1)

plt.xscale('symlog')

#plt.axis([0, 33554432, 0, 100])
plt.ylabel("accuracy (%)")  




#plt.style.use('ggplot')

fig = plt.figure(figsize=(15, 4))
ax = fig.add_subplot(111)
ax.set_xscale("log")
#ax.axvline(524288, color='gray', ls='--', label="0%")
ax.vlines(x = [524288], ymin= 0, ymax= 100, color='lightgray', ls='--', label="0'%' acc loss on 524,288 bits flip")
ax.vlines(x = [8388608], ymin= 0, ymax= 100, color='black', ls='--', label="1'%' acc loss on 8,388,608 bits flip")
ax.vlines(x = [20000000], ymin= 0, ymax= 100, color='teal', ls='--', label="5'%' acc loss on 20,000,000 bits flip")
ax.vlines(x = [23000000], ymin= 0, ymax= 100, color='orange', ls='--', label="10'%' acc loss on 23,000,000 bits flip")
ax.plot(x, y, 'bo-', label='Accuracy', linewidth=1)
ax.set_xlabel("Number of bit flips")
ax.set_ylabel("accuracy (%)")
ax.set_title("BER sensitivity")
plt.legend()



x_ber = []

for i in range (len(x)):
    x_ber.append((x[i]/47185919))

fig_a = plt.figure(figsize=(15, 4))
ax_a = fig_a.add_subplot(111)
ax_a.set_xscale("log")
ax_a.vlines(x = [524288/47185919], ymin= 0, ymax= 100, color='lightgray', ls='--', label="0%")
ax_a.vlines(x = [8388608/47185919], ymin= 0, ymax= 100, color='black', ls='--', label="1%")
ax_a.vlines(x = [20000000/47185919], ymin= 0, ymax= 100, color='teal', ls='--', label="5%")
ax_a.vlines(x = [23000000/47185919], ymin= 0, ymax= 100, color='orange', ls='--', label="10%")
ax_a.plot(x_ber, y, 'ro-', label='Accuracy', linewidth=1)
ax_a.set_xlabel("Bit Error Rate")
ax_a.set_ylabel("accuracy (%)")
ax_a.set_title("BER sensitivity")
plt.legend()



x_a = []

for i in range (len(x)):
    x_a.append((x[i]/47185919)*100)

fig_a = plt.figure(figsize=(15, 4))
ax_b = fig_a.add_subplot(111)
ax_b.set_xscale("linear")
ax_b.vlines(x = [(524288/47185919)*100], ymin= 0, ymax= 100, color='lightgray', ls='--', label="0'%' acc loss at 1.11'%' of bits")
ax_b.vlines(x = [(8388608/47185919)*100], ymin= 0, ymax= 100, color='black', ls='--', label="1'%' acc loss at 17.77'%' of bits")
ax_b.vlines(x = [(20000000/47185919)*100], ymin= 0, ymax= 100, color='teal', ls='--', label="5'%' acc loss at 42.38'%' of bits")
ax_b.vlines(x = [(23000000/47185919)*100], ymin= 0, ymax= 100, color='orange', ls='--', label="10'%' acc loss at 48.74'%' of bits")
ax_b.plot(x_a, y, 'go-', label='Accuracy', linewidth=1)
ax_b.set_xlabel("Portion of bit flips (%)")
ax_b.set_ylabel("accuracy (%)")
ax_b.set_title("BER sensitivity")
plt.legend()

plt.show()
```
